# Mushroom Classification

<b>This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

<b>In this project we need to predict the "Class" column.</b>

# Dataset Information

<b>The data set contains 8124 rows and the following 23 features:</b>

<b>class:</b> edible(e) or poisonous(p)

<b>cap-shape:</b> bell(b), conical(c), convex(x), flat(f), knobbed(k), sunken(s)

<b>bruises:</b> bruises(t), no bruises(f)

<b>gill-attachment:</b> attached(a), descending(d), free(f), notched(n)

<b>gill-spacing:</b> close(c), crowded(w), distant(d)

<b>gill-size:</b> broad(b), narrow(n)

<b>gill-color:</b> black(k), brown(n), buff(b), chocolate(h), gray(g), green(r), orange(o), pink(p), purple(u), red(e), white(w), yellow(y)

<b>stalk-shape:</b> enlarging(e), tapering(t)

<b>stalk-root:</b> bulbous(b), club(c), cup(u), equal(e), rhizomorphs(z), rooted(r), missing(?)

<b>stalk-surface-above-ring:</b> fibrous(f), scaly(y), silky(k), smooth(s)

<b>stalk-surface-below-ring:</b> fibrous(f), scaly(y), silky(k), smooth(s)

<b>stalk-color-above-ring:</b> brown(n), buff(b), cinnamon(c), gray(g), orange(o), pink(p), red(e), white(w), yellow(y)

<b>stalk-color-below-ring:</b> brown(n), buff(b), cinnamon(c), gray(g), orange(o), pink(p), red(e), white(w), yellow(y)

<b>veil-type:</b> partial(p), universal(u)

<b>veil-color:</b> brown(n), orange(o), white(w), yellow(y)

<b>ring-number:</b> none(n), one(o), two(t)

<b>ring-type:</b> cobwebby(c), evanescent(e), flaring(f), large(l), none(n), pendant(p), sheathing(s), zone(z)

<b>spore-print-color:</b> black(k), brown(n), buff(b), tan(h), green(r), orange(o), purple(u), white(w), yellow(y)

<b>population:</b> abundant(a), clustered(c), numerous(n), scattered(s), several(v), solitary(y)

# Data Manipulation

<b>The data is categorical so we’ll use Label Encoder to convert it to ordinal. Label Encoder converts each value in a column to a number.</b>

<b>This approach requires the category column to be of ‘category’ datatype. By default, a non-numerical column is of ‘object’ datatype. From the df.describe() method, we saw that our columns are of ‘object’ datatype. So we will have to change the type to ‘category’ before using this approach.</b>

```python
df = df.astype('category')
```

```python
df.dtypes
```

```python
# Changing the categorical varaiables to numerical variables

from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
for column in df.columns:
    df[column] = LE.fit_transform(df[column])
```

<b>Observations:</b>

<b>Cap Shape:</b> There are certainly not very many characteristics of cap shape that could definitively decide if the mushroom is poisonous or edible. If the mushroom is knobbed, more often than not it could be poisonous. A bell shaped cap is more likely to be edible.

<b>Cap Surface:</b> If the mushroom has a more fibrous surface, it’s more likely to be edible. Smooth and scaly mushrooms are slightly more likely to be poisonous.

<b>Cap Color:</b> If the mushroom has a more white or gray surface, it’s likely to be edible. Red or yellow cap colors tend to be more poisonous.

<b>Bruises:</b> If the mushroom has bruises, it’s likely to be edible. If the mushroom does not have bruises, it’s more likely to be poisonous.

<b>Odor:</b> If the mushroom has no odor, it’s extremely likely to be edible. Any odor that is foul or pungent is highly likely to be poisonous.

<b>Gill Attachments:</b> There are no characteristics that could definitively classify the mushroom as edible or poisonous.

<b>Gill Spacing:</b> If the gills are closely spaced, it’s like to be poisonous. If they are crowded, then it’s more likely to be edible.

<b>Gill Size:</b> If the gills are narrow, it’s like to be poisonous. If they are broad, then it’s more likely to be edible.

<b>Gill Color:</b> Colors of buff are almost certain to be poisonous. Gray and chocolate colors are also likely to be poisonous. White, purple, and brown colors are likely to be edible.

<b>Stalk Shape:</b> Stalks are difficult to analyze for edibility.

<b>Stalk Root:</b> Missing roots are very likely to be poisonous. Club and equal roots are likely to be edible.

<b>Stalk Surface Above Ring:</b> Smooth stalks above the ring will likely be edible. Silky stalks are likely to be poisonous.

<b>Stalk Surface Below Ring:</b> Smooth stalks above the ring will likely be edible. Silky stalks are likely to be poisonous.

<b>Stalk Color Above Ring:</b> White and gray colors are likely to be edible. Buff, brown, and purple are likely to be poisonous.

<b>Stalk Color Below Ring:</b> White and gray colors are likely to be edible. Buff, brown, and purple are likely to be poisonous.

<b>Veil Color:</b> Veils are difficult to analyze for edibility.

<b>Ring Number:</b> Ring Numbers are difficult to analyze for edibility.

<b>Ring Type:</b> Pendant ring types are very likely to be edible. Evanescent and large ring types are likely to be poisonous.

<b>Spore Print Color:</b> Black and brown are highly likely to be edible. Tan and white colors are highly likely to be poisonous.

<b>Population:</b> Several mushrooms found in the population are likely to be poisonous. Numerous and abundant are likely to be edible.

<b>Habitat:</b> Mushrooms found in grasses and woody areas are likely to be edible. Mushrooms found on paths and leaves are likely to be poisonous.

# Separating the features and label

```python
x = df.drop('class', axis=1)
y = df['class']
```

# Finding the best random state

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

maxAccu=0
maxRS=0

for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    DTC = DecisionTreeClassifier()
    DTC.fit(x_train, y_train)
    pred = DTC.predict(x_test)
    acc=accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```

# Building Machine Learning Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
```

# Logistic Regression

```python
# Checking Accuracy and evaluation metrics for Logistic Regression
LR = LogisticRegression()

# Training the model
LR.fit(x_train,y_train)

#Predicting y_test
predLR = LR.predict(x_test)

# Accuracy Score
LR_score = accuracy_score(y_test, predLR)*100
print("Accuracy Score:", LR_score)

# ROC AUC Score
from sklearn.metrics import roc_auc_score
roc_auc_score = roc_auc_score(y_test,predLR)*100      
print("\nroc_auc_score:", roc_auc_score)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, predLR)
print("\nConfusion Matrix:\n",conf_matrix)

# Classification Report
class_report = classification_report(y_test,predLR)
print("\nClassification Report:\n", class_report)

# Cross Validation Score
cv_score = (cross_val_score(LR, x, y, cv=5).mean())*100
print("Cross Validation Score:", cv_score)

# Result of accuracy minus cv scores
Result = LR_score - cv_score
print("\nAccuracy Score - Cross Validation Score is", Result)
```

```python
# Lets plot confusion matrix for Logistic Regression

cm = confusion_matrix(y_test,predLR)
x_axis_labels = ["0","1"]
y_axis_labels = ["0","1"]

f , ax = plt.subplots(figsize=(5,4))
sns.heatmap(cm, annot = True,linewidths=.2, linecolor="black",fmt = ".0f", ax=ax, cmap="coolwarm",xticklabels=x_axis_labels,yticklabels=y_axis_labels)

plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for Logistic Regression')
plt.show()
```

# Decision Tree Classifier

```python
# Checking Accuracy and evaluation metrics for Decision Tree Classifier
DTC = DecisionTreeClassifier()

# Training the model
DTC.fit(x_train,y_train)

#Predicting y_test
predDTC = DTC.predict(x_test)

# Accuracy Score
DTC_score = accuracy_score(y_test, predDTC)*100
print("Accuracy Score:", DTC_score)

# ROC AUC Score
from sklearn.metrics import roc_auc_score
roc_auc_score = roc_auc_score(y_test,predDTC)*100      
print("\nroc_auc_score:", roc_auc_score)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, predDTC)
print("\nConfusion Matrix:\n",conf_matrix)

# Classification Report
class_report = classification_report(y_test,predDTC)
print("\nClassification Report:\n", class_report)

# Cross Validation Score
cv_score = (cross_val_score(DTC, x, y, cv=5).mean())*100
print("Cross Validation Score:", cv_score)

# Result of accuracy minus cv scores
Result = DTC_score - cv_score
print("\nAccuracy Score - Cross Validation Score is", Result)
```

```python
# Lets plot confusion matrix for Decision Tree Classifier

cm = confusion_matrix(y_test,predDTC)
x_axis_labels = ["0","1"]
y_axis_labels = ["0","1"]

f , ax = plt.subplots(figsize=(5,4))
sns.heatmap(cm, annot = True,linewidths=.2, linecolor="black",fmt = ".0f", ax=ax, cmap="coolwarm",xticklabels=x_axis_labels,yticklabels=y_axis_labels)

plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.show()
```
